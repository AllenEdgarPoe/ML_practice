```python
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer = load_breast_cancer()
print('cancer.keys(): \n{}'.format(cancer.keys()))
print('data shape: {}'.format(cancer.data.shape))

X = cancer.data
y= cancer.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
training_accuracy = []
test_accuracy = []
for n in range(1, 11): 
    classifier = KNeighborsClassifier(n)
    classifier.fit(X_train, y_train)
    training_accuracy.append(np.mean(classifier.predict(X_train)==y_train))
    test_accuracy.append(np.mean(classifier.predict(X_test)==y_test))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot(training_accuracy, label = 'train accuracy')
plt.plot(test_accuracy, label = 'test accuracy')
plt.xlabel('Accuracy')
plt.legend()
```
